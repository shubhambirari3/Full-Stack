```python
#another exapmle

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_person_info(self):
        print(f"Name: {self.name}, Age: {self.age}")

class Education:
    def __init__(self, degree, university):
        self.degree = degree
        self.university = university

    def display_education_info(self):
        print(f"Degree: {self.degree}, University: {self.university}")

class Student(Person, Education):
    def __init__(self, name, age, degree, university, student_id):
        # Initialize both parent classes
        Person.__init__(self, name, age)
        Education.__init__(self, degree, university)
        self.student_id = student_id

    def display_student_info(self):
        self.display_person_info()
        self.display_education_info()
        print(f"Student ID: {self.student_id}")

```

```python
class Grandparent:
    def __init__(self, family_name):
        self.family_name = family_name

    def show_family_name(self):
        print(f"Family Name: {self.family_name}")

class Parent(Grandparent):
    def __init__(self, family_name, parent_name):
        super().__init__(family_name)
        self.parent_name = parent_name

    def show_parent_name(self):
        print(f"Parent Name: {self.parent_name}")

class Child(Parent):
    def __init__(self, family_name, parent_name, child_name):
        super().__init__(family_name, parent_name)
        self.child_name = child_name

    def show_child_name(self):
        print(f"Child Name: {self.child_name}")


```

```python
# Creating an instance of Child
child = Child("Smith", "John Smith", "Emma Smith")

# Accessing methods from all levels of inheritance
child.show_family_name()  # From Grandparent class
child.show_parent_name()  # From Parent class
child.show_child_name()   # From Child class
```

```python
# Parent Class
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print(f"{self.name} makes a sound.")

# Child Class 1
class Dog(Animal):
    def speak(self):
        print(f"{self.name} barks.")

# Child Class 2
class Cat(Animal):
    def speak(self):
        print(f"{self.name} meows.")

# Child Class 3
class Cow(Animal):
    def speak(self):
        print(f"{self.name} moos.")

```

```python
# Creating Instances of Each Child Class
dog = Dog("sheru")
cat = Cat("mau")
cow = Cow("gayya")

# Calling their 'speak' Methods
dog.speak()  # Output: Buddy barks.
cat.speak()  # Output: Whiskers meows.
cow.speak()  # Output: Daisy moos.
```

```python
# Parent Class
class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def show_info(self):
        print(f"Brand: {self.brand}, Model: {self.model}")

# Child Class 1
class Car(Vehicle):
    def __init__(self, brand, model, doors):
        super().__init__(brand, model)
        self.doors = doors

    def show_info(self):
        super().show_info()
        print(f"Car with {self.doors} doors.")

# Child Class 2
class Motorcycle(Vehicle):
    def __init__(self, brand, model, engine_capacity):
        super().__init__(brand, model)
        self.engine_capacity = engine_capacity

    def show_info(self):
        super().show_info()
        print(f"Motorcycle with {self.engine_capacity}cc engine.")

# Child Class 3
class Truck(Vehicle):
    def __init__(self, brand, model, load_capacity):
        super().__init__(brand, model)
        self.load_capacity = load_capacity

    def show_info(self):
        super().show_info()
        print(f"Truck with {self.load_capacity} tons load capacity.")


```

```python
# Creating Instances of Each Child Class
car = Car("Toyota", "Corolla", 4)
motorcycle = Motorcycle("Yamaha", "MT-15", 150)
truck = Truck("Volvo", "FH16", 20)

# Displaying Information
car.show_info()
print()  # Separator
motorcycle.show_info()
print()  # Separator
truck.show_info()
```

What is Polymorphism in Python?

Polymorphism means "many forms" and refers to the ability of different objects to use the same interface (method or behavior) in different ways.
It allows objects of different classes to respond to the same method name in their own way.
Polymorphism enhances code reusability and flexibility.

Types of Polymorphism in Python:

Method Overriding: A subclass provides a specific implementation of a method already defined in its parent class.

Method Overloading: Python does not directly support method overloading, but you can achieve it using default arguments or *args.

Polymorphism with Functions and Classes: A function can take different types of objects as arguments and call the same method on them.

### What is Encapsulation?
Encapsulation is one of the fundamental principles of Object-Oriented Programming (OOP). It is the process of bundling data (attributes) and the methods (functions) that operate on the data into a single unit, typically a class. It also involves restricting direct access to some of the object's components to ensure data security and proper control.

```python
class BankAccount:
    def __init__(self, account_number, balance):
        self.__account_number = account_number  # Private attribute
        self.__balance = balance  # Private attribute

    # Public method to access private data
    def get_balance(self):
        return self.__balance

    # Public method to update private data
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited {amount}. New balance: {self.__balance}")
        else:
            print("Invalid deposit amount")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew {amount}. Remaining balance: {self.__balance}")
        else:
            print("Invalid withdrawal amount")

# Create a BankAccount object
account = BankAccount("123456789", 1000)

# Accessing balance using public method
print("Balance:", account.get_balance())  # Output: Balance: 1000

# Deposit money
account.deposit(500)  # Output: Deposited 500. New balance: 1500

# Withdraw money
account.withdraw(200)  # Output: Withdrew 200. Remaining balance: 1300

# Trying to access private attributes directly (not allowed)
# print(account.__balance)  # AttributeError: 'BankAccount' object has no attribute '__balance'
```

# abstraction


```python
from abc import ABC, abstractmethod

class Shape(ABC):  # Abstract class
    @abstractmethod
    def area(self):  # Abstract method
        pass

    @abstractmethod
    def perimeter(self):  # Abstract method
        pass

class Rectangle(Shape):  # Concrete subclass
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)

class Circle(Shape):  # Another subclass
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

    def perimeter(self):
        return 2 * 3.14 * self.radius

# Instantiate subclasses and use their methods
rect = Rectangle(10, 5)
print("Rectangle Area:", rect.area())  # Output: 50
print("Rectangle Perimeter:", rect.perimeter())  # Output: 30

circle = Circle(7)
print("Circle Area:", circle.area())  # Output: 153.86
print("Circle Perimeter:", circle.perimeter())  # Output: 43.96
```

What Is Abstraction?
Abstraction is a principle in Object-Oriented Programming (OOP) that focuses on hiding implementation details and exposing only the essential features of an object.
It allows us to define what a class should do, but not how it does it. The "how" part is left to the subclasses.
Abstraction is achieved using abstract classes and abstract methods.

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):  # Abstract class
    @abstractmethod
    def start_engine(self):  # Abstract method
        pass

    @abstractmethod
    def stop_engine(self):  # Abstract method
        pass

class Car(Vehicle):  # Concrete subclass
    def start_engine(self):
        return "Car engine started."

    def stop_engine(self):
        return "Car engine stopped."

class Bike(Vehicle):  # Another concrete subclass
    def start_engine(self):
        return "Bike engine started."

    def stop_engine(self):
        return "Bike engine stopped."

# Instantiate and use concrete subclasses
car = Car()
print(car.start_engine())  # Output: Car engine started.
print(car.stop_engine())   # Output: Car engine stopped.

bike = Bike()
print(bike.start_engine())  # Output: Bike engine started.
print(bike.stop_engine())   # Output: Bike engine stopped.
```
