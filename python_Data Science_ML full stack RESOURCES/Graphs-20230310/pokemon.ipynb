```python
df = pd.read_csv('E:\Python\Machine Learning\datasets\Pokemon.csv')
```

```python
obj=df.groupby("Type 1")
```

```python
avg_df = obj.mean()
```

```python
avg_df
```

```python
import matplotlib.pyplot as plt
```

```python
avg_df.index
```

```python
list(zip(avg_df.index, avg_df['Attack'],avg_df['Speed']))
```

```python
plt.figure(figsize=(16, 10))
plt.scatter(avg_df['Attack'],avg_df['Speed'])

for label,a,s in zip(avg_df.index, avg_df['Attack'],avg_df['Speed']):
    plt.annotate(label,
                 xy=(a, s), # put the label with its point
                 xytext=(4, 5),                  # but slightly offset
                 textcoords='offset points')
plt.title("type 1 average attack against avg speed")
plt.xlabel("avg attack")
plt.ylabel("avg Speed")
plt.show()
```
