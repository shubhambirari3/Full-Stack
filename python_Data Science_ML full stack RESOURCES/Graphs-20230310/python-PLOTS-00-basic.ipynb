#### Basic Visualization Rules

1. __Type of plot__ : The first step is to choose the appropriate plot type. 
2. __Axes formation__ : Second, when we choose type of plot, one of the most important things is to label your axis. If we don’t do this, the plot is not informative enough. 
3. __Title__ : we can add a title to make our plot more informative.
4. __Labels__: add labels for different categories when needed.
5. __Annotation, optionally we can add a text or an arrow at interesting data points.
6. __ Figure size, color etc__ in some cases we can use some sizes and colors of the data to make the plot more informative.

#### Basic matplotlib

> - import matplotlib
> - import matplotlib.pyplot as plt
> - from matplotlib import style
> - %matplotlib inline 

- __plt.title__(“My Title”) will add a title “My Title” to your plot
- __plt.xlabel__(“Year”) will add a label “Year” to your x-axis
- __plt.ylabel__(“Population”) will add a label “Population” to your y-axis
- __plt.xticks__([1, 2, 3, 4, 5]) set the numbers on the x-axis to be 1, 2, 3, 4, 5. We can also pass and labels as a second argument. For, example, if we use this code plt.xticks([1, 2, 3, 4, 5], ["1M", "2M", "3M", "4M", "5M"]), it will set the labels 1M, 2M, 3M, 4M, 5M on the x-axis.
- __plt.yticks__() - works the same as plt.xticks(), but for the y-axis.

#### Histograms

- is an accurate representation of the distribution of numeric data. 
- To create a histogram, first, we divide the entire range of values into a series of intervals, and second, we count how many values fall into each interval. 
- The intervals are also called bins. The bins are consecutive and non-overlapping intervals of a variable. They must be adjacent and are often of equal size.
- To make a histogram with Matplotlib, we can use the plt.hist() function. The first argument is the numeric data, the second argument is the number of bins. 
- The default value for the bins argument is 10.

```python
numbers = [0.1, 0.1,0.5, 1, 1.5, 2, 4, 5.5, 6, 8, 8.5,9, 9.9]

plt.hist(numbers, 4)

plt.xlabel("Number")
plt.ylabel("Frequency")
plt.show()
```

```python
style.use('classic')

# Create a figure instance
fig = plt.figure(1, figsize=(9, 6))

# Create an axes instance
ax = fig.add_subplot(111)

# Create the boxplot
#bp = ax.boxplot(data_to_plot)
bp = ax.boxplot(data_to_plot, patch_artist=True)
#print(bp)
## change outline color, fill color and linewidth of the boxes
for box in bp['boxes']:
    # change outline color
    box.set( color='g', linewidth=2)
    # change fill color
    box.set( facecolor = 'y' )
## change color and linewidth of the whiskers
for whisker in bp['whiskers']:
    whisker.set(color='r', linewidth=2)
## change color and linewidth of the caps
for cap in bp['caps']:
    cap.set(color='b', linewidth=2)

## change color and linewidth of the medians
for median in bp['medians']:
    median.set(color='b', linewidth=2)

## change the style of fliers and their fill
for flier in bp['fliers']:
    flier.set(marker='o', color='#e7298a', alpha=0.9)

ax.set_xticklabels(['Sample1', 'Sample2', 'Sample3', 'Sample4'])
```
