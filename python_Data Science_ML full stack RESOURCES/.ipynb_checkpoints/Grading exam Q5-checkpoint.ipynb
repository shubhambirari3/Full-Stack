#### check selection count of type1 as grass and type1 as  fire is same or not

```python
from scipy.stats import chi2_contingency

# Create a contingency table for Type 1 Grass and Type 1 Fire Pokemon selection counts
grass_fire_ct = pd.crosstab(df['Type 1'], df['Legendary'])

# Perform chi-squared test of independence to compare selection counts for Type 1 Grass and Type 1 Fire Pokemon
chi2_statistic, p_value, dof, expected = chi2_contingency(grass_fire_ct)
print("chi2_statistic :" , chi2_statistic)
print("p_value" , p_value)

# Print the results of the chi-squared test
if p_value < 0.05:
    print("The selection count for Type 1 Grass and Type 1 Fire Pokemon is statistically different (p-value: {0:.4f}).".format(p_value))
else:
    print("The selection count for Type 1 Grass and Type 1 Fire Pokemon is not statistically different (p-value: {0:.4f}).".format(p_value))
```

#### check if speed of type1 as fire and type1 as fire with type2 as Dragon same or not

```python
import pandas as pd
from scipy.stats import ttest_ind

# Extract Type 1 Fire and Type 1 Fire with Type 2 Dragon Pokemon data
fire_df = df[(df['Type 1'] == 'Fire') & ((df['Type 2'] == 'Dragon') | (df['Type 2'].isnull()))]
speed_fire = fire_df['Speed'].values

# Perform two-sample t-test to compare speed of Type 1 Fire and Type 1 Fire with Type 2 Dragon Pokemon
t_statistic, p_value = ttest_ind(speed_fire, df[df['Type 1'] == 'Fire']['Speed'].values, equal_var=False)
print('t_statistic :' , t_statistic)
print('p_value :' , p_value)

# Print the results of the t-test
if p_value < 0.05:
    print("The speed of Type 1 Fire and Type 1 Fire with Type 2 Dragon Pokemon is statistically different (p-value: {0:.4f}).".format(p_value))
else:
    print("The speed of Type 1 Fire and Type 1 Fire with Type 2 Dragon Pokemon is not statistically different (p-value: {0:.4f}).".format(p_value))
```
