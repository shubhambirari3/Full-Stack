```python
#line plot
#whenever time component is involved on x axis
#depict the relation between two variables in the trend


x = [1, 2, 3, 4, 5]
y = [5, 1, 7, 8, 2]

plt.plot(x, y)
# Creates a line plot with `x` and `y` data points.
```

```python
plt.plot(x, y, color = "red", marker = "o", linestyle = '--') #dotted line
plt.title("line plot to understand the syntax") #title of the plot
plt.xlabel("x points") #x axis label
plt.ylabel("y points")  #y axis label
plt.show()  #display the plot
```

```python
plt.plot(x, y, color = "red", marker = "o", linestyle = '--', linewidth = 3)
plt.title("line plot to understand the syntax")
plt.xlabel("x points")
plt.ylabel("y points")
plt.show()
```

```python
plt.plot(x, y, color = "red", marker = "o", linestyle = '--', linewidth = 3, markersize=10)
plt.title("line plot to understand the syntax")
plt.xlabel("x points")
plt.ylabel("y points")
plt.show()
```

```python
plt.plot(x, y, color = "red", marker = "o", linestyle = '--', linewidth = 3, markersize=10) #dotted line with red color  and circle marker with line width 3 and marker size 10
plt.title("line plot to understand the syntax")
plt.xlabel("x points")
plt.ylabel("y points")
plt.grid()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [5, 1, 7, 8, 2]
y2 = [7, 8, 3, 5, 1]


plt.plot(x, y1, color = "red", marker = "o", linestyle = '--', linewidth = 3, markersize=6, label = 'line y1')
plt.plot(x, y2, color = "b", marker = "o", linestyle = '--', linewidth = 3, markersize=10, label = 'line y2')

plt.xlabel("x data axis")
plt.ylabel("y data axis")

plt.title("Multiple lines in one plot")
plt.legend()

plt.show()
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
# Creates two line plots: one for `sin(x)` and one for `cos(x)` using the `x` array.
```

```python
plt.plot(x, np.sin(x), '--')
# Creates a dashed line plot for `sin(x)` using the `x` array.
```

```python
plt.plot(x, np.sin(x), '-')
# Creates a solid line plot for `sin(x)` using the `x` array.
```

```python
plt.plot(x, np.sin(x), '.-') #dotted line with circle marker for sin(x) using x array 
# why we use sin  and cos? to understand the trend of the data
```

```python
plt.plot(x, np.sin(x), ':')
```

```python
plt.figure() # Creates a new figure. why we need this? to create a new plot
plt.subplot(1, 2, 1) #subplot 1 row, 2 columns, 1st plot
plt.plot(x, np.sin(x)) #subplot 1


plt.subplot(1, 2, 2) #subplot 1 row, 2 columns, 2nd plot
plt.plot(x, np.cos(x)) #subplot 2

plt.show()
# Creates a figure with two subplots in a 1x2 grid. 
# The first subplot (left) plots `sin(x)`,
# and the second subplot (right) plots `cos(x)`.
```

```python
plt.figure() 
plt.subplot(2, 2, 1) #subplot 2 rows, 2 columns, 1st plot
plt.plot(x, np.sin(x)) #subplot 1 of 4 plots in 2x2 grid 


plt.subplot(2, 2, 2) #subplot 2 rows, 2 columns, 2nd plot
plt.plot(x, np.cos(x)) #subplot 2 of 4 plots in 2x2 grid

plt.subplot(2, 2, 3) #subplot 2 rows, 2 columns, 3rd plot
plt.plot(x, np.tan(x)) #subplot 3 of 4 plots in 2x2 grid 

plt.subplot(2, 2, 4) #subplot 2 rows, 2 columns, 4th plot 
plt.plot(x, np.cos(x+1)) #subplot 4 of 4 plots in 2x2 grid

plt.show() #display the plot with 4 subplots in 2x2 grid 
```

```python
plt.figure()    # Creates a new figure. 
plt.subplot(2, 1, 1) #subplot 2 rows, 1 column, 1st plot 
plt.plot(x, np.sin(x)) #subplot 1 of 2 plots in 2x1 grid


plt.subplot(2, 1, 2) #subplot 2 rows, 1 column, 2nd plot 
plt.plot(x, np.cos(x)) #subplot 2 of 2 plots in 2x1 grid


plt.show() #display the plot with 2 subplots in 2x1 grid 
# Creates a figure with two subplots in a 2x1 grid. The first subplot (top) plots `sin(x)`, and the second subplot (bottom) plots `cos(x)`.
```

```python
#pie chart >> To understand the percnt dist of data

langs = ['C', 'Python', 'Java']
students = [20, 100, 40]


plt.pie(students, labels = langs, autopct = '%1.1f%%') #pie chart for students with labels and percentage upto 1 decimal point
plt.title("Pie chart") #title of the plot 
plt.show() #display the plot
```

```python
explode = (0.0, 0.1, 0.1)

langs = ['C', 'Python', 'Java']
students = [20, 100, 40]


plt.pie(students, labels = langs, autopct = '%1.1f%%', explode = explode, shadow = True) #pie chart for students with labels, percentage upto 1 decimal point, explode and shadow effect  
plt.title("Pie chart")
plt.show()
```

```python
#3d plot

x = np.random.rand(20)
y = np.random.rand(20)
z = np.random.rand(20)


fig = plt.figure() # Creates a new figure  
ax = fig.add_subplot(projection = '3d') #3d plot   what is add_subplot? to add a subplot to the current figure
ax.scatter(x, y, z, c='red') #red color 3d scatter plot for x, y and z data points
```

```python
import matplotlib.pyplot as plt

# Sample data
years = [2021, 2018, 2022, 2023, 2017]
category1 = [20, 25, 30, 28, 35]
category2 = [10, 15, 20, 18, 25]
category3 = [5, 8, 10, 9, 15]

# Create a area plot
plt.stackplot(years, category1, category2, category3, labels=['Category 1', 'Category 2', 'Category 3'],
              colors=['skyblue', 'lightgreen', 'orange'])

# Add labels and title
plt.xlabel("Years")
plt.ylabel("Values")
plt.title("Area Plot")
plt.legend()

plt.show()
```
