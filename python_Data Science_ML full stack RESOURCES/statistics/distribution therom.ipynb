```python
from scipy.stats import binom
```

```python
## k is number of success
## n is total number of matches
## p= Probability to win
s= binom.pmf(k=3,n=5,p=0.75)
s
```

```python

t= binom.pmf(k=4,n=5,p=0.75)
t
```

```python
q= binom.pmf(k=5,n=5,p=0.75)
q
```

```python
s+t+q
```

```python
## do the same using for loop
tot=0
for i in range(3,6):
    q= binom.pmf(k=i,n=5,p=0.75)
    print(q)
    tot=q+tot
print(tot)
```
