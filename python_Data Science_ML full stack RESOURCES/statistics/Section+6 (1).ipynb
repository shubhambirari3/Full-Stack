## Central Limit Theorem

Two aspects of central limit theorem:

    1)for almost all populations, the sampling distribution of the mean can be approximated closely by normal distribution, Provided sample size is sufficiently large.

    2)If the variable has a mean of mu and variance as var , as the sample size n increases,  the sample mean approaches a normal distribution with mean as mu(x) and variance std(x**2)
    
    The observation is as the number of items for creating sample mean is increases sandard deviation will be narrower.
    
    Standard deviation of the sampling distribution of the sample mean is called Standanrd error

### Hypothesis Testing Using Python

Commonly used tests include:

    compare sample statistic with the population parameter

    compare two datasets
    
    Summary of sample is called as statistics and summary of population is called as parameter.
    
    based on sample statistics we want to conclude about the population parameter.


##### A machine is producing perfume bottles with the long term average of 150cc and the standard deviation of 2 cc. Four bottles were picked and the average volume was found to be 153.8cc. Has the average volume changed from 150cc? Test with 95% confidence level.

## Steps to perform Hypothesis Testing:
    
    1) State the alternate hypothesis
    
    2) State the null hypothesis
    
    3) select the probability error level (alpha level) generally 0.05
    
    4) calculate the test statistic (e.g. t-score or z- score)
    
    5)critical test statistics
    
    6)Inteprete the result
    

Alternate Hypothesis: Mean volume != 150cc

Null Hypothesis: Mean volume = 150cc

Alpha = 1 - 0.95 = 0.05

Z calculated = (Observed mean - population mean) / standard error


z = (x  ) / ( / n) 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Z_cal = (153.8 - 150) / (2/np.sqrt(4))
# import math
# Z_cal = (153.8 - 150) / (2/math.sqrt(4))
```

```python
Z_cal
```

```python
# norm.pdf(x, loc=0, scale=1) Probability density function.

# norm.cdf(x, loc=0, scale=1)  Cumulative distribution function. (area to the left of x)

# norm.sf(x, loc=0, scale=1) Survival function (also defined as 1 - cdf, but sf is sometimes more accurate).

# norm.ppf(q, loc=0, scale=1) Percent point function (inverse of cdf  percentiles).

# norm.isf(q, loc=0, scale=1) Inverse survival function (inverse of sf).
```

```python
from scipy.stats import norm 
# we learnt this earlier when discussing the Normal Distribution
```

```python
norm.cdf(1) 
```

```python
norm.cdf(-1.96)
```

```python
norm.sf(1)
```

```python
norm.sf(1.96)
```

```python
Z_crit_left = norm.ppf(0.025)
Z_crit_left
```

```python
Z_crit_right = norm.isf(0.025)
Z_crit_right
```

```python
#Conclusion: Reject the Null Hypothesis
```

```python
# Thinking in terms of p-value
```

```python
pvalue = 2 * norm.sf(Z_cal) # area to the right of Z_cal (on two tails)
pvalue
```

```python
Compared to the alpha value of 0.05, the p-value is low, hence null gets rejected.
```
