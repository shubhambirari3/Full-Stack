```python
df = pd.DataFrame({'A':[1,2,np.nan],
                  'B':[5,np.nan,np.nan],
                  'C':[1,2,3]})
```

```python
df
```

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

```python
df
```

```python
df.dropna(thresh=2, axis =1)
```

```python
df.columns
```

```python
df.fillna('ABC')
```

```python
df.columns
```

```python
for b in df.columns:
    print(b)
    df[b]=df[b].fillna(df[b].mean())
```

```python
df
```

```python
df.fillna(value=df.mean())
```
