```python
location = r"E:\Python\Machine Learning\datasets\iris.csv"
```

```python
# load the training data from glass data set
df = pd.read_csv(location)
```

```python
df.head()
```

```python
df['class'].value_counts()
```

```python
# Label Encoding
df['class'].replace(["Iris-setosa","Iris-versicolor", "Iris-virginica"], [2,1,0], inplace=True)
```

```python
df.head()
```

```python
X = df.drop(['class'], axis=1).values
y = df['class'].values
```

```python
# Create logistic regression
logistic = LogisticRegression()
```

```python
#from sklearn.linear_model import Log
```

```python
import numpy as np
# Create regularization penalty space
penalty = ['l1', 'l2']

# Create regularization hyperparameter space
C = np.logspace(0, 3, 10)
print(C)

# Create hyperparameter options
hyperparameters = dict(C=C, penalty=penalty)
```

```python
hyperparameters
```

```python
import sklearn
sklearn.__version__
```

```python
# Create grid search using 5-fold cross validation
clf = GridSearchCV(logistic, hyperparameters, cv=5,)
```

```python
# Fit grid search
best_model = clf.fit(X, y)
```

```python
# View best hyperparameters
print('Best Penalty:', best_model.best_estimator_.get_params()['penalty'])
print('Best C:', best_model.best_estimator_.get_params()['C'])
```

```python
logistic = LogisticRegression(C=12.91)
```

```python
logistic.fit(X,y)
```

```python
y_pred= logistic.predict(X)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y,y_pred)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y,y_pred))
```
