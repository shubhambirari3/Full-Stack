```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

plt.figure(figsize=(12, 12))
```

### make_blobs

__n_samples__ : int or array-like, optional (default=100)
If int, it is the total number of points equally divided among clusters. If array-like, each element of the sequence indicates the number of samples per cluster.

__n_features__ : int, optional (default=2)
The number of features for each sample.

__centers__ : (default=None) The number of centers to generate, or the fixed center locations. If n_samples is an int and centers is None, 3 centers are generated. 

__cluster_std__ : float or sequence of floats, optional (default=1.0)
The standard deviation of the clusters.


```python
n_samples    = 1500
random_state = 170

X, y = make_blobs(n_samples=n_samples, random_state=random_state)
```

```python
X.shape
```

```python
np.unique(y)
```

```python
# Incorrect number of clusters
y_pred = KMeans(n_clusters=3, random_state=random_state).fit_predict(X)

#plt.subplot(221)
plt.scatter(X[:, 0], 
            X[:, 1], 
            c=y_pred)

plt.title("Incorrect Number of Blobs")
```

```python
# distributed data
transformation = [[0.60834549, -0.63667341], [-0.40887718, 0.85253229]]
X_aniso        = np.dot(X, transformation)

y_pred  = KMeans(n_clusters=3, random_state=random_state).fit_predict(X_aniso)

#plt.subplot(222)
plt.scatter(X_aniso[:, 0], 
            X_aniso[:, 1], 
            c=y_pred)

plt.title("Anisotropicly Distributed Blobs")
```
