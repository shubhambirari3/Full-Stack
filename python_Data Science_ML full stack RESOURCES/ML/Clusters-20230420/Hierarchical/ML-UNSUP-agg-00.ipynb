```python
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
```

```python
np.random.seed(123)
```

```python

x = np.random.multivariate_normal([10, 0], [[3, 1], [1, 4]],
                                  size=[10,])
x
```

```python

y = np.random.multivariate_normal([0, 20], [[3, 1], [1, 4]], 
                                  size=[20,])
y
```

```python
# generating two clusters: x with 10 points and y with 20:

X = np.concatenate((x, y))
print(X)
```

```python
print(X.shape)  # 150 samples with 2 dimensions
plt.scatter(X[:, 0], X[:, 1])

plt.show()
```

```python
y.shape
```

```python
X.shape
```

```python
# generate the linkage matrix
Z = linkage(X)
#print(Z)
```

```python
from scipy.cluster.hierarchy import cophenet
from scipy.spatial.distance import pdist
coph_dists = cophenet(Z, pdist(X))
#coph_dists
```

```python
plt.figure(figsize=(25, 13))
plt.title('HCA Dendrogram')
plt.xlabel('sample index')
plt.ylabel('distance')
dendrogram(Z,leaf_rotation=90,leaf_font_size=12,)
plt.show()
```
