```python
#Null Hypothesis: attack done by type 1 water and type 1 Ice is same
#Alternative Hypothesis: attack done by type 1 water and type 1 Ice is not same

water=df[df["Type 1"]=='Water']["Attack"]
ice=df[df["Type 1"]=='Ice']["Attack"]
A=weightstats.ztest(water,ice,alternative='two-sided')
print("z-test score", A[0])
print("p-value", A[1])
```

```python
#Null Hypothesis: selection count of type1 as grass and type1 as fire is same
#Alternative Hypothesis: selection count of type1 as grass and type1 as fire is not same

from statsmodels.stats.proportion import proportions_ztest
grass_count = df.loc[df["Type 1"] == "Grass"].shape[0]
fire_count = df.loc[df["Type 1"] == "Fire"].shape[0]
print(grass_count,fire_count)
n1 = 800
n2 = 800
count1 = grass_count
count2 = fire_count
counts = [count1, count2]
nobs = [n1, n2]
stat, pval = proportions_ztest(counts, nobs)
print("z-statistic:", stat)
print("p-value:", pval)
```

```python
#Null Hypothesis: Speed of type1 as fire and type1 as fire with type2 as Dragon is same

#Alternative Hypothesis: Speed of type1 as fire and type1 as fire with type2 as Dragon is not same

T1_fire_T2_Dragon = df[(df['Type 1']=='Fire') & (df['Type 2']=='Dragon')]['Speed']
T1_fire = df[(df['Type 1']=='Fire')]['Speed']
from scipy.stats import f_oneway
t_stat, p_value = f_oneway(T1_fire,T1_fire_T2_Dragon)
print('Test statistics: ',t_stat)
print('P value: ',p_value)
```
