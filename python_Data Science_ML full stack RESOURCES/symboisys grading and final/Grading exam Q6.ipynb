```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\shubh\Downloads\Pokemon.csv")
```

```python
df
```

#### Generate a histogram showing the distribution of Pokemon total stats:

```python
sns.set_style('whitegrid')
sns.set_palette('bright')
plt.figure(figsize=(10,8))

sns.histplot(df['Total'], bins=50)
plt.title('Distribution of Pokemon Total Stats', fontsize=18)
plt.xlabel('Total', fontsize=14)
plt.ylabel('Count', fontsize=14)

plt.show()
```

#### Generate a box plot showing the distribution of Pokemon total stats by type:

```python
sns.set_style('whitegrid')
sns.set_palette('bright')
plt.figure(figsize=(12,6))

sns.boxplot(x='Type 1', y='Total', data=df)
plt.xticks(rotation=90)
plt.title('Distribution of Pokemon Total Stats by Type', fontsize=18)
plt.xlabel('Type 1', fontsize=14)
plt.ylabel('Total', fontsize=14)

plt.show()
```

#### Generate a stacked bar plot showing the count of legendary and non-legendary Pokemon by type:

```python
legendary_count = df.groupby(['Type 1', 'Legendary']).size().unstack()
legendary_count.plot(kind='bar', stacked=True, figsize=(12,6))

plt.title('Count of Legendary and Non-Legendary Pokemon by Type', fontsize=18)
plt.xlabel('Type 1', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=90)

plt.show()
```

#### Generate a heatmap showing the correlation between different Pokemon stats:

```python
sns.set_style('white')
sns.set_palette('bright')
plt.figure(figsize=(10,8))

corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.title('Correlation Between Pokemon Stats', fontsize=18)

plt.show()
```
