#### a)	Create the basket for discount on Monthly MB:                                                                                 

```python

def get_monthly_mb_discount(monthly_mb):
    if monthly_mb <= 50:
        return 0.1
    elif monthly_mb <= 1000:
        return 0.04
    elif monthly_mb <= 15000:
        return 0.03
    elif monthly_mb <= 30000:
        return 0.02
    else:
        return 0.0

# Add a new column to the merged dataset with the discount percentages
combine_df['monthly_mb_discount'] = combine_df['monthly_mb'].apply(get_monthly_mb_discount)

```

#### b) Create the new column “Message” having message for user as mentioned below

```python
def generate_message(monthly_mb):
    bonus_mb = 0
    
    if monthly_mb <= 50:
        bonus_mb = 5
    elif monthly_mb <= 1000:
        bonus_mb = monthly_mb * 0.1
    elif monthly_mb <= 15000:
        bonus_mb = monthly_mb * 0.03
    elif monthly_mb <= 30000:
        bonus_mb = monthly_mb * 0.02
    
    total_mb = monthly_mb + bonus_mb
    
    return f"Greetings, Your Monthly used data is {monthly_mb}MB. You will get {total_mb}MB data to use for the next month."

combine_df['Message'] = combine_df['monthly_mb'].apply(generate_message)

combine_df
```

#### c) Find

```python
#i) Which platform user uses highest Monthly_mb used

#Group by platform and calculate mean monthly_mb
platform_usage = combine_df.groupby('platform').mean()['monthly_mb']
sorted_platform_usage = platform_usage.sort_values(ascending=False)
highest_platform = sorted_platform_usage.index[0]

print(f"The platform with the highest mean monthly_mb is {highest_platform}")
```

```python
#ii) Least used device

device_usage = combine_df.groupby('device').sum()['monthly_mb']
sorted_device_usage = device_usage.sort_values()
least_used_device = sorted_device_usage.index[0]

print(f"The least used device is {least_used_device}")
```

```python
#iii)Count of the IOS users

ios_users = combine_df['platform'].value_counts()['ios']

print(f"The number of IOS users is {ios_users}")
```
