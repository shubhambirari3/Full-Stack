### Generate any 10 grapahs suitable for Superstore data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df  = pd.read_excel(r"C:\Users\shubh\symbiosis ml16\pandas\Superstore.xls")
```

```python
df
```

```python
#Bar plot of total sales by region:
sales_by_region = df.groupby('Region')['Sales'].sum()
sales_by_region.plot(kind='bar', title='Total Sales by Region')
plt.show()
```

```python
#Pie chart of total sales by category:
sales_by_category = df.groupby('Category')['Sales'].sum()
sales_by_category.plot(kind='pie', title='Total Sales by Category')
plt.show()
```

```python
#Line plot of profit over time:
profit_over_time = df.groupby('Order Date')['Profit'].sum()
profit_over_time.plot(kind='line', title='Profit over Time')
plt.show()
```

```python
#tacked bar plot of sales and profit by category:
sales_and_profit_by_category = df.groupby('Category')[['Sales', 'Profit']].sum()
sales_and_profit_by_category.plot(kind='bar', stacked=True, title='Sales and Profit by Category')
plt.show()
```

```python
#Scatter plot of sales and profit:
df.plot(kind='scatter', x='Sales', y='Profit', title='Sales vs. Profit')
plt.show()
```

```python
#Box plot of sales by category:
df.boxplot(column='Sales', by='Category')
plt.title('')
plt.suptitle('Sales by Category')
plt.show()
```

```python
#Histogram of sales:
df['Sales'].plot(kind='hist', bins=20, title='Sales Distribution')
plt.show()
```

```python
#Area plot of sales by category over time:
sales_by_category_and_time = df.groupby(['Category', 'Order Date'])['Sales'].sum().unstack('Category')
sales_by_category_and_time.plot(kind='area', title='Sales by Category over Time')
plt.show()
```

```python
pd =df.groupby('Sales')['Profit'].sum()
pd.plot(kind = 'area' , figsize = (20,10), alpha = .4,stacked = False)
```

```python
#Violin plot of sales by region:

sns.violinplot(x='Region', y='Sales', data=df)
plt.title('Sales by Region')
plt.show()
```
