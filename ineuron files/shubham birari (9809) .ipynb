```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('bmh')
def f(x,a,b):
    return(b/a)*((x/a)**(b-1))
xlist=np.linspace(0.4,1.7,num=1000)
ylist1=f(xlist,2,0.5)
ylist2=f(xlist,2,1)
ylist3=f(xlist,2,1.5)
plt. figure(num=0, dpi=120)
plt.plot(xlist,ylist1,label="f(x1),beta<1")
plt.plot(xlist,ylist2,'--r',label="f(x2), beta=1" )
plt.plot(xlist,ylist3,'--g',label="f(x3), beta>1")
plt.figtext(.3,.2,'IFR CLASS',c='g')
plt.figtext(.3,.6,"DFR class",c='b')
plt.figtext(0.2,0.4, "constant failure rate class", c='r')
plt.ylabel("Hazard rate r(t)")
plt.xlabel("time (t)")
plt.title("Hazard rate of weibull distribution")
plt.legend();plt.show()
```

```python
def f(x,a):
    return 1+(x*a)
xlist=np.linspace(0.4,1.7,num=1000)
ylist1=f(xlist,0.5)
ylist2=f(xlist,1)
ylist3=f(xlist,2)
plt.figure(num=0,dpi=120)
plt.plot(xlist,ylist1,label="f(x1),theta1")
plt.plot(xlist,ylist2,'--r',label="f(x2),theta=1")
plt.plot(xlist,ylist3,'--g',label="f(3),theta1")
plt.figtext(.6,.2,'IFR class',c='b')
plt.figtext(.6,.8,"IFR class",c='g')
plt.figtext(.6,.5,"IFR class",c='r')
plt.ylabel ("Hazard rate r(t)")
plt.xlabel("time(t)")
plt.title ("Hazard rate of linear failure rate family")
plt.legend();plt.show()
```

```python
def f(x,a):
    return 1+(a* (1-np.exp(-x)))
xlist=np.linspace(0.4,1.7,num=1000)
ylist1=f(xlist,0.5)
ylist2=f(xlist,1)
ylist3=f(xlist,2)
plt.figure(num=0,dpi=120)
plt.plot(xlist,ylist1,label="f(x1),theta<1")
plt.plot(xlist,ylist2,'--r',label="f(x2),theta=1")
plt.plot(xlist,ylist3,'--g',label="f(3),theta1")
plt.figtext(.6,.2,'IFR class',c='b')
plt.figtext(.6,.8,'IFR class',c='g')
plt.figtext(.6,.5,"IFR CLASS",c='r')
plt.ylabel("Hazard rate r(t)")
plt.xlabel("time(t)")
plt.title("Hazard rate of makeham familiy")
plt.legend();plt.show()
```

```python
import scipy.stats as st
def f(x, sigma, mu):
    cnst=1/(sigma*x*np.sqrt(2*np.pi))
    t2=np.exp(-0.5*((np.log(x)-mu)/sigma)**2)
    fx=cnst*t2
    sx=1-st.norm.cdf((np.log(x)-mu)/sigma)
    return fx/sx
xlist=np.linspace(0.4,1.7, num=1000)
ylistl=f(xlist, 0.4,2) 
ylist2=f(xlist,0.6,2)
ylist3=f(xlist, 0.8,2)
ylist4=f(xlist,1,2)
plt.figure(num=0, dpi=120)
plt.plot(xlist,ylist1,label="f(x1), sigma-0.4, mu=2") 
plt.plot(xlist,ylist2,'--r',label="f(x2), sigma=0.6, mu=2")
plt.plot(xlist,ylist3,'--g',label="f(x3), sigma=0.8, mu=2") 
plt.plot(xlist,ylist4,'--y',label="f(x4), sigma=1, mu=2")
plt.figtext(.2,.7,'Constant failure rate class',c='b') 
plt.figtext(.2,.65,"IFR CLASS",c="y") 
plt.figtext(.2,.6,"IFR CLASS",c='g')
plt.figtext(.2,.55,"IFR CLASS",c='r')
plt.ylabel("Hazard rate r(t)")
plt.xlabel("tiime(t)") 
plt.title("Hazard rate of lognormal distribution")
plt.legend();plt.show()

```
